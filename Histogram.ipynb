```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

# Plotting simple Histogram

```python
ml_students_age = np.random.randint(18,45, (100))  # age is between 18 -25 and need 100 records
py_students_age = np.random.randint(15,40, (100))
plt.hist(ml_students_age)
plt.show()
```

# Title and labels

```python
plt.hist(ml_students_age)
 
plt.title("ML Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.show()
```

# Creating Bins and multiple parameters

```python
bins = [15,20,25,30,35,40,45] # category of ML students age on x axis
plt.figure(figsize = (16,9)) # size of histogram in 16:9 format
 
                                                              # histtype means the filled bar
plt.hist(ml_students_age, bins, rwidth=0.8, histtype = "bar",  # rwidth 0.8 means show bar 80% and 20% show gap
         orientation='vertical', color = "m", label = "ML Student")  # orientation means display horizontal or vertical
 
plt.title("ML Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.legend()
plt.show()
```

# Histogram with multi colum bars

```python

from matplotlib import style # for style 
style.use("ggplot") # return grid
plt.figure(figsize = (16,9)) 
 
plt.hist([ml_students_age, py_students_age], bins, rwidth=0.8, histtype = "bar",
         orientation='vertical', color = ["m", "y"], label = ["ML Student", "Py Student"])
 
# gime both the parameters in the list because histogram only accept one argument and give list of colors and list of labels
plt.title("ML & Py Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.legend()
plt.show()
```
